```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = r'C:\Users\9812170414089\ADS-Assignment-2-3\WA_Fn-UseC_-HR-Employee-Attrition.csv'
df= pd.read_csv(path)
```

```python
df.head()
```

```python
grouped_data = df.groupby(['JobRole', 'Attrition'])['DistanceFromHome'].value_counts().reset_index(name='count')
grouped_data
```

```python
sns.catplot(x='DistanceFromHome', y='count', hue='Attrition', col='JobRole', kind='point', data=grouped_data)
plt.show()
```

```python
df.groupby(['Education', 'Attrition']).agg({'MonthlyIncome':['mean']})
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 6))
df.groupby(['Education', 'Attrition']).agg({'MonthlyIncome': ['mean']}).unstack().plot(kind='bar', ax=ax)

# Set plot properties
ax.set_xlabel('Education Level')
ax.set_ylabel('Mean Monthly Income')
ax.set_title('Mean Monthly Income by Education Level and Attrition')
ax.legend(title='Attrition', labels=['No', 'Yes'])
plt.tight_layout()

# Show the plot
plt.show()
```
